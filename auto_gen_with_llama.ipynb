```python
from autogen import AssistantAgent
import autogen

# Setup API key. Add your own API key to config file or environment variable
config_list = [
    {
        "model": "llama3.2:3b",
        "base_url": "http://localhost:11434/v1",
        'api_key': 'ollama',
    },
]

llm_config = {
    "request_timeout": 60,
    "seed": 42,
    "config_list": config_list,
    "temperature": 0
}

assistant = AssistantAgent(
    name="small model",
    max_consecutive_auto_reply=2,
    system_message="""You are an expert in Power Studio SCADA by CIRCUTOR. You can answer all technical questions about SCADA, 
        analyze electrical data, provide analysis of images, and assist with creating screens or dashboards. 
        You can also communicate with SCADA API to check device connections, provide real-time device information, 
        and assist in understanding how the entire SCADA system works.""",
    llm_config={
        "config_list": config_list,
        "temperature": 0,
    }
)

big = autogen.UserProxyAgent(
    name="user_proxy",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    code_execution_config={"work_dir": "web"},  # Corrected to dict with proper key-value
    llm_config=llm_config,
    system_message="""
    Reply TERMINATE if the task has been solved to full satisfaction.
    Otherwise, reply CONTINUE or explain why the task is not solved yet.
    """
)

# Task to initiate
task = """explique moi la ligne de base selon iso5001"""

# Initiate the conversation
big.initiate_chat(assistant, message=task)
```
